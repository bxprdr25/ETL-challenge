```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# NYPD Complaint Data
complaint_file = "../Resources/NYPD_Complaint_Data_2017.csv"

NYPD_data =pd.read_csv(complaint_file)

NYPD_data.head()
```

```python
NYPD_data.count()
```

```python
NYPD_df = NYPD_data

NYPD_df
```

```python
NYPD_df["CMPLNT_FR_DT"].value_counts() 
```

```python
# Get the year 2017
startdate = "1/1/2017"
enddate = "2/28/2017"

after_start_date_DT = NYPD_df["CMPLNT_FR_DT"] >= startdate
before_end_date_DT = NYPD_df["CMPLNT_FR_DT"] <= enddate
between_two_dates = after_start_date_DT & before_end_date_DT
NYPD_2017_df = NYPD_df[between_two_dates]

NYPD_2017_df
```

```python
NYPD_2017_new_df = NYPD_2017_df.rename(columns={'CMPLNT_NUM' : 'Complaint_number_ID', 'BORO_NM' : 'BORO_Name', 'CMPLNT_FR_DT': 'Start_Date', 'CMPLNT_TO_DT': 'End_Date', 'CRM_ATPT_CPTD_CD': 'Complaint_Status','KY_CD':'Complaint_code', 'LAW_CAT_CD':'Complaint_CAT', 'SUSP_AGE_GROUP' :'Suspect_Age', 'SUSP_RACE' : 'Suspect_Race', 'SUSP_SEX' : 'Suspect_Gender', 'VIC_AGE_GROUP' : 'Victim_Age', 'VIC_RACE' : 'Victim_Race','VIC_SEX' : 'Victim_Gender', 'OFNS_DESC' : 'Complaint_DESC'  })

NYPD_2017_new_df
```

```python
NYPD_2017_new_df.dtypes
```

```python
NYPD_2017_new_df["Start_Date"] = pd.to_datetime(NYPD_2017_new_df["Start_Date"], format="%m/%d/%Y", errors="coerce")

NYPD_2017_new_df.info()
```

```python
NYPD_2017_2m_df= NYPD_2017_new_df[(NYPD_2017_new_df["Start_Date"].dt.month_name() == "January") | (NYPD_2017_new_df["Start_Date"].dt.month_name() == "February")]

NYPD_2017_2m_df
```

```python
# DropNA
NYPD_2017_2m_df= NYPD_2017_2m_df.dropna()

NYPD_2017_2m_df
```
